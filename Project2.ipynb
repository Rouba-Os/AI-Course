```python
# Load Cleane Stock Data From Project1
import pandas as pd

df = pd.read_csv('cleaned_stock_data_project1.csv', parse_dates=['date'], index_col='date')

print("Dataset Preview ")
print(df.head())

print("\nDataset Info")
print(df.info())
```

```python
# Check Missing Values
print("Missing Values Check")
print(df.isnull().sum())
```

```python
# Detect Outliers (IQR Method)
Q1 = df['close'].quantile(0.25)
Q3 = df['close'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[(df['close'] < lower) | (df['close'] > upper)]

print("Number of Outliers (Close)")
print(len(outliers))
```

```python
# =========================
# Handle Outliers (Capping)
# =========================
df['close'] = df['close'].clip(lower, upper)

print("After Capping Outliers (Close)")
print(df['close'].describe())
```

```python
# ===============
# Error Detection
# ================
print("Check for Negative Values")
print(df[df['close'] < 0])
print(df[df['volume'] < 0])
```

```python
# ============
# Fix errors
# ============
df = df[df['close'] > 0]
df = df[df['volume'] > 0]

print("After Removing Invalid Values")
print(df.shape)
```

```python
Feature Engineering
## Moving Averages
df['MA_10'] = df['close'].rolling(window=10).mean()
df['MA_30'] = df['close'].rolling(window=30).mean()

## Volatility
df['Volatility'] = df['close'].rolling(window=10).std()

## Daily Returns
df['Returns'] = df['close'].pct_change()
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[['open','high','low','close','volume']] = scaler.fit_transform(
    df[['open','high','low','close','volume']]
)
```

```python
# Encode Categorical Variables
df = pd.get_dummies(df, columns=['sector'], drop_first=True)

print("Dataset After Encoding Sector")
print(df.head())
```

```python
# Data Splitting (Time Series Split )
train = df[:'2015']
val = df['2016':'2018']
test = df['2019':]
```

```python
# Save Final Data
df.to_csv('final_cleaned_data.csv')
train.to_csv('train.csv')
val.to_csv('val.csv')
test.to_csv('test.csv')

print("Files Saved Successfully")
```
