```python
# Import libraries
import pandas as pd
import numpy as np
from functools import reduce
```

```python
# Task 1: Create a DataFrame using Pandas
df = pd.DataFrame({"Name": ["Alice", "Bob", "Charlie", "David"],
                   "Age": [25, 30, 35, 28],
                   "City": ["New York", "San Francisco", "Los Angeles", "Chicago"]})

print("Task 1 Initial DataFrame:")
print(df)
```

```python
# Task 2: Row and Column Manipulation
# Drop the 'City' column from the DataFrame created in Task 1
df_dropped = df.drop(columns=["City"])
print("Task 2 DataFrame after dropping 'City':")
print(df_dropped)
```

```python
# Task 3: Handling Null Values
# Create a new DataFrame with some null values and fill them
df_null = pd.DataFrame({"Name": ["Alice", None, "Charlie", "David"],
                        "Age": [25, np.nan, 35, 28],
                        "City": ["New York", "San Francisco", None, "Chicago"]})

df_filled = df_null.fillna({"Name": "Unknown",
                            "Age": 0,
                            "City": "Unknown"})

print("Task 3 DataFrame after filling nulls:")
print(df_filled)
```

```python
# Task 4: GroupBy and Describe
# Using the following DataFrame, group by 'Category' and describe the 'Value' column
df_group = pd.DataFrame({
    'Category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Value': [10, 20, 15, 25, 30, 35]
})

grouped_stats = df_group.groupby("Category")["Value"].describe()
print("Task 4 GroupBy and Describe:")
print(grouped_stats)
```

```python
# Task 5: Concatenation and Merging
# Concatenate two DataFrames vertically and merge them horizontally

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})
df3 = pd.DataFrame({'C': [9, 10], 'D': [11, 12]})

concat_df = pd.concat([df1, df2], ignore_index=True)

merged_df = pd.concat([concat_df, df3], axis=1)

print("Task 5 Concatenated DataFrame:")
print(concat_df)
print("\nTask 5 Merged DataFrame:")
print(merged_df)
```

```python
# Task 6: Tuples and Sets
# Create a tuple of fruits and a set of numbers.
# Then, try to add an element to each and observe the difference.

# Tuple of fruits
fruits = ("apple", "banana", "cherry")

# Set of numbers
numbers = {1, 2, 3, 4, 5}

# Add new elemnt to the tuple of fruits
fruits.add("orange")

# Create a new tuple with the added fruit
fruits = fruits + ("orange",)

#  Add new elemnt to the set of numbers
numbers.add(6)

print("Task 6 Tuple:", fruits)
print("Task 6 Set:", numbers)

print("Tuples are immutable (cannot be changed), so we create a new tuple instead.")
print("Sets are mutable (can be updated directly).")
```

```python
# Task 8: Functions and Lambda
# Create a function to calculate the square of a number.
# Then, use a lambda function to do the same.

## Regular function
def square(x):
    return x**2

## Lambda function
square_lambda = lambda x: x**2

print("Task 8 Square of 3 using function:", square(3))
print("Task 8 Square of 6 using lambda:", square_lambda(6))
```

```python
# Task 10: Map, Reduce, and Filter
# Use map to square all numbers in a list.
# Use reduce to find the product of all numbers in a list.
# Use filter to get only even numbers from a list.

numbers = [1, 2, 3, 4, 5]

# Map to square all numbers
squared = list(map(lambda x: x**2, numbers))

# Reduce to find the product of all numbers
product = reduce(lambda x, y: x*y, numbers)

# Filter to get only even numbers
evens = list(filter(lambda x: x%2 == 0, numbers))

print("\nTask 10 Squared numbers:", squared)
print("Task 10 Product of numbers:", product)
print("Task 10 Even numbers:", evens)
```

```python
# Task 11: Object-Oriented Programming - Creating a Class
# Create a class called 'Rectangle' with attributes 'length' and 'width'.
# Include methods to calculate area and perimeter.

class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    def area(self):
        return self.length * self.width
    def perimeter(self):
        return 2*(self.length + self.width)

rect1 = Rectangle(4, 5)

print("Task 11 Rectangle 1 area:", rect1.area())
print("Task 11 Rectangle 1 perimeter:", rect1.perimeter())
```

```python
# Task 12: Pandas Data Analysis
df_employees = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Charlie'],
    'Department': ['IT', 'HR', 'IT', 'Finance', 'HR'],
    'Salary': [55000, 65000, 70000, 60000, 58000]
})

# 1. Find the average salary by department
avg_salary = df_employees.groupby("Department")["Salary"].mean()
print("Task 12 Average salary by department:")
print(avg_salary)

# 2. Get the names of employees with salary > 60000
high_salary_employees = df_employees[df_employees["Salary"] > 60000]["Name"]
print("\nTask 12 Employees with salary > 60000:")
print(high_salary_employees.tolist())

# 3. Add a new column 'Bonus' which is 10% of salary
df_employees["Bonus"] = df_employees["Salary"] * 0.1
print("\nTask 12 DataFrame with Bonus column:")
print(df_employees)
```
