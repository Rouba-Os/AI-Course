# GPTs architecture and functionality

Architecture:
It breaks text into small pieces called tokens
Converts them into numbers (embeddings)
Uses transformer layers with attention to understand context
Predicts the next word using an output layer

Functionality:
It is trained on large amounts of text (pre-training)
Then it generates text by predicting one word at a time based on what came before

```python
# Install the necessary libraries and import the required modules
!pip install transformers torch

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# Load the pre-trained model and tokenizer
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
```

```python
# Define function for text generation
def generate_text(prompt, max_length=120, temperature=1.0, top_k=40, top_p=0.9):

    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True
    )

    return tokenizer.decode(output[0], skip_special_tokens=True)
```

```python
# Experiment with different prompts and parameters
## Basic Generation
prompt = "Summer time is"
print("Basic text generation")
print(generate_text(prompt))
```

```python
## Controlling output length
prompt = "Artificial intelligence will change"
print("Controlling output length")
print(generate_text(prompt, max_length=60))
print(generate_text(prompt, max_length=140))
```

```python
## Temperature Effect
prompt = "business intelligence in healthcare sector will improve"

print("Low temperature:")
print(generate_text(prompt, temperature=0.3))

print("\nHigh temperature:")
print(generate_text(prompt, temperature=1.4))
```

```python
## Using top_k and top_p
prompt = "Future jobs will require skills in"

print("Using top_k and top_p")
print("Default settings:")
print(generate_text(prompt))
print("\ntop_k and top_p:")
print(generate_text(prompt, top_k=20, top_p=0.7))
```

```python
## Prompt engineering
print("Prompt engineering")
base_prompt = "Write a short story about"
topics = ["a self-driving car making decisions", "how automation is changing the workplace", "a robot working in a restaurant"]

for topic in topics:
    full_prompt = f"{base_prompt} {topic}"
    print(f"Topic: {topic}")
    print(generate_text(full_prompt, max_length=140))
    print("\n" + "-"*30 + "\n")
```

```python
# Application example
def article_generator(topic):
    return generate_text(f"Write a short article about {topic}:", max_length=180)

print("Application example")
print(article_generator("AI in finance"))
```
